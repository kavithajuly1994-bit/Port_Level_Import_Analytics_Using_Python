## Port Performance Classification

```python
port_value = df.groupby('port')['Import_Value_Inr'].transform('sum')

df['Port_Category'] = pd.qcut(
    port_value,
    q=3,
    labels=['Low Performing','Medium Performing','High Performing']
)
```

```python
df['Quantity_Reported'] = np.where(df['quantity'] > 0, 'Yes', 'No')
```

* Two derived fields were created to enhance business analysis.
* Port_Category classifies ports into Low, Medium, and High Performing groups based on their total import value contribution, helping identify major trade hubs.
* Quantity_Reported was created to distinguish transactions with available quantity information from those where quantity was not reported, enabling data quality.

```python
df.columns
```

```python
# Removel of unwanted spaces
df['country'] = df['country'].str.strip()
```

```python
# Port column
df['port'] = df['port'].str.strip()
```

```python
# Commodity Column
df['commodity'] = df['commodity'].str.strip()
```

Removing unwanted spaces ensured that identical countries, ports, and commodities were grouped correctly, resulting in more accurate import statistics and dashboard visualizations.

```python
# Date Column
df['date'] = pd.to_datetime(df['date'])
```

```python
# Extract Year
df['year'] = df['date'].dt.year

# Extract Month Name
df['month'] = df['date'].dt.month_name()

# Extract Quarter
df['Quarter'] = df['date'].dt.quarter
```

* The Date column was converted into datetime format to enable time-based analysis.
* Additional features such as Year, Month, and Quarter were derived from the Date column to support annual, monthly, and quarterly trend analysis.

```python
# Currency Formatting
df['Import_Value_Inr'] = df['Import_Value_Inr'].round(2)

df['Usd_Value'] = df['Usd_Value'].round(2)
```

* Standardized monetary values to two decimal places.
* Financial values became easier to interpret in reports and dashboards.

```python
# Handling Missing value
df['units'] = df['units'].fillna('Unknown')
```

* Missing values were present only in the Units column.
* Missing unit information was replaced with "Unknown".
* "Unknown" accurately represents unavailable information.

## **Predictive Analysis: Import Trends and Future Trade Direction**

## **Monthly Import Trend**

```python
monthly_year = df.groupby(['year','month'])['Import_Value_Inr'].sum().reset_index()

month_order = [
    'January','February','March','April',
    'May','June','July','August',
    'September','October','November','December']

monthly_year['month'] = pd.Categorical(
    monthly_year['month'],
    categories=month_order,
    ordered=True)

monthly_year = monthly_year.sort_values(['year','month'])

plt.figure(figsize=(12,6))

sns.lineplot(
    data=monthly_year,
    x='month',
    y='Import_Value_Inr',
    hue='year',
    marker='o')

plt.title('Monthly Import Trend (2023 vs 2024)')
plt.xlabel('Month')
plt.ylabel('Import Value (INR)')
plt.xticks(rotation=45)

plt.show()
```

The chart shows how import values changed month-by-month and compares import performance between 2023 and 2024.

**Features Used**
* Used Month, Year, and Import_Value_Inr.
* Used sns.lineplot() with hue='year' and marker='o'.
* Compared monthly import trends between 2023 and 2024.

**Final Insights **

* Import value increased steadily from **January** (4.33 trillion INR) to **March** (5.02 trillion INR) in 2023.
* The highest import value in 2023 was recorded in **March** (5.02 trillion INR).
* In **2024**, imports increased sharply from January (4.48 trillion INR) to February (5.05 trillion INR).
* **February 2024** recorded the highest import value across all months shown.
Import value declined from 5.05 trillion INR in February to 4.73 trillion INR in March 2024.
* 2024 imports were higher than 2023 in **January** and **February**.
